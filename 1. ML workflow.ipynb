#### Defining Machine Learning: What can ML do?

Can you predict a social media comment is toxic or non-toxic? **Classification**
<br>Can you predict whether or not a customer will purchase a particular item? **Classification**
<br>Can you predict a positive or negative sentiment to a given restaurant review? **Classification**
<br>Can you predict whether an email is a spam or not? **Classification**
<br>Can you predict a tumor is benign or malignant? **Classification**
<br>Can you predict whether the value of a stock will go up or go down next week? **Classification**
<br>Can you predict the price of a new house based on various features? **Regression**
<br>Can you predict what will be the value of stock next week? **Regression**
<br>With the given time and weather data predict the number of bikes being rented at any given hour? **Regression**
<br>Can you group the students of an online education company into different categories based on their learning styles? **Clustering**
<br>Find groups of customers with similar behavior given a large database of customer data containing their properties and past buying records. **Clustering**
<br>Is this a picture of a cat or a dog? **Classification**
<br>Is it going to rain or not? **Classification**
<br>Is this borrower going to repay their loan? **Classification**
<br>What is the genre of this song/movie? **Classification**


#### How ML works?
- Interdisciplinary mix of statistics and computer science
- Ability to learn without being explicitly programmed
- Learn patterns from existing data and apply it to new dat
- Relies on high-quality data etc.

#### Terminologies
- Training data, historic data
- Model
- Features
- Target
- Observations
- Test data, unseen data
- Accuracy metric


I. **Supervised Learning**
- Training data is **labeled**

1. **Regression**
- Linear Regression
- Support Vector Machine
- Decision Tree
- Random Forest
- Gradient Boosting
- Xtreme Gradient Boosting

2. **Classification**
- Logistic Regression
- Support Vector Machine
- Decision Tree
- Random Forest
- Gradient Boosting
- Xtreme Gradient Boosting

II. **Unsupervised Learning**
- Training data has only features, there is no target variable
- Clustering (K-means, DBScan, Hierarchical clustering etc), Dimensionality Reduction (Principal Component Analysis), Recommendation (Non-Negative Matrix Factorization)

III. **Deep Learning**
- Convolutional Neural Network (CNN)
- Long Short-Term memory (LSTM)
